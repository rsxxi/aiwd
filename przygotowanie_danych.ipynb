## PrzeksztaÅ‚canie typÃ³w zmiennych

```python
#Sprawdzamy typy zmiennych
data.dtypes
```

```python
'''Wszystkie dane majÄ… typ object, na pewno warto wiÄ™c przeksztaÅ‚ciÄ‡ typ czÄ™Å›ci z nich

Po pierwsze, zmienne Region i School Type przeksztaÅ‚cimy na typy category (sÄ… to zmienne kategorialne)

Po drugie, wszystkie zmienne zwiÄ…zane z wymagrodzeniami chcemy przeksztaÅ‚ciÄ‡ w zmienne numeryczne,
bÄ™dzie to wymagaÅ‚o rÃ³wnieÅ¼ usuniÄ™cia z nich znakÃ³w '$' i ','
'''
```

```python
# Nadajemy typ zmiennym kategorialnym
data['School Type'] = data['School Type'].astype('category')
data['Region'] = data['Region'].astype('category')
```

```python
# Zmieniammy typ danych numerycznych

# Wczytujemy listÄ™ kolumn ze zmniennymi numerycznymi
numerical=list(data.columns[3:])

# Oczyszczamy zawartoÅ›Ä‡ zmiennych z niepoÅ¼Ä…danych znakÃ³w i zmieniamy ich typ
for n in numerical:
    data[n] = data[n].str.replace(',', '')
    data[n] = data[n].str.replace('$', '')
    data[n] = data[n].astype(float)
```

## Braki danych

```python
# Sprawdzamy jakich kolumn dotyczÄ… braki i ile ich jest
columns=list(data.columns)
missing=[]
for col in columns: missing.append({col, data[col].isnull().sum()})
missing
```

```python
#Jak widzimy, jeÅ›li w rekordzie wystÄ™pujÄ… braki danych, to sÄ… one obecne dla obu kolumn

#NajwiÄ™cej brakÃ³w danych dotyczy kategorii Liberal Arts dla typu uniwersytetu
#Dla innych typÃ³w sÄ… to braki jednostkowe
#Sprawdzamy wiÄ™c, jak wiele jest brakujÄ…cych danych w stosunku do danych peÅ‚nych dla tego typu
len(data.loc[data['School Type'] == 'Liberal Arts'].index)
```

NajrozsÄ…dniejszym wyjÅ›ciem wydaje siÄ™ wykluczenie z analiz z udziaÅ‚em zmiennych o 10 i 90 centylu zarobkÃ³w wszystkich uniwersytetÃ³w z kategorii Liberal Arts lub przeprowadzanie ich jedynie na 13 peÅ‚nych rekordach z wyraÅºnym zaznaczeniem brakÃ³w. PozostaÅ‚e braki dotyczÄ… kategorii Engineering oraz Party, sÄ… na tyle nieliczne, Å¼e ich uzupeÅ‚nienie nie powinno zaburzyÄ‡ znaczÄ…co analiz.

## UzupeÅ‚nienie brakÃ³w danych (z wyÅ‚Ä…czeniem Liberal Arts)

```python
'''Iterujemy kolejno po kolumnach w ktÃ³rych wystÄ™pujÄ… braki danych
UÅ¼ywamy funkcji .loc, by wybraÄ‡ jedynie rekordy, ktÃ³rych typem nie jest "Liberal Arts"
UÅ¼ywamy funkcji .groupby, by wypeÅ‚niÄ‡ brakujÄ…ce dane Å›rednimi dla danego typu, uÅ¼ywamy iteratora jako
indeksu wskazujÄ…cego na odpowiedniÄ… kolumnÄ™
W danej kolumnie uzupeÅ‚niamy kaÅ¼dy brak danych Å›redniÄ… uÅ¼ywajÄ…c funkcji .mean
'''

for _ in ['Mid-Career 10th Percentile Salary', 'Mid-Career 90th Percentile Salary']:
    data.loc[data['School Type'] != 'Liberal Arts',_] = \
    data.groupby('School Type')[_].apply(lambda x:x.fillna(x.mean()))
```
