## Opis zmiennych

**School Name** - zmienna typu nominalnego, unikatowe nazwy poszczególnych uniwersytetów.  

**Region** - zmienna kategorialna typu nominalnego, region, w którym zlokalizowany jest uniwersytet.  

**School Type** - zmienna kategorialna typu nominalnego, typ (profil) uniwersytetu.  

**Starting Median Salary** - zmienna numeryczna typu ilorazowego. Mediana początkowych zarobków absolwentów uniwersytetu.  

**Mid-Career Median Salary** - zmienna numeryczna typu ilorazowego. Mediana zarobków absolwentów uniwersytetu w trakcie trwania ich kariery (10 lat po ukończeniu studiów).  

Pozostałe zmienne to zmienne numeryczne typu ilarozowego. Są to zarobki odpowiednio dla 10, 25, 75 i 90 centyla (pod kątem wysokości wypłaty) absolwentów uniwersytetu w trakcie trwania ich kariery (10 lat po ukończeniu studiów)

## Przekształcanie typów zmiennych

```python
#Sprawdzamy typy zmiennych
data.dtypes
```

```python
'''Wszystkie dane mają typ object, na pewno warto więc przekształcić typ części z nich

Po pierwsze, zmienne Region i School Type przekształcimy na typy category (są to zmienne kategorialne)

Po drugie, wszystkie zmienne związane z wymagrodzeniami chcemy przekształcić w zmienne numeryczne,
będzie to wymagało również usunięcia z nich znaków '$' i ','
'''
```

```python
# Nadajemy typ zmiennym kategorialnym
data['School Type'] = data['School Type'].astype('category')
data['Region'] = data['Region'].astype('category')
```

```python
# Zmieniammy typ danych numerycznych

# Wczytujemy listę kolumn ze zmniennymi numerycznymi
numerical=list(data.columns[3:])

# Oczyszczamy zawartość zmiennych z niepożądanych znaków i zmieniamy ich typ
for n in numerical:
    data[n] = data[n].str.replace(',', '')
    data[n] = data[n].str.replace('$', '')
    data[n] = data[n].astype(float)
```

```python
# Sprawdzamy, czy dane numeryczne zostały poprawnie zmienione
data
```

```python
# Sprawdzamy typy zmiennych
data.dtypes
```

## Braki danych

```python
# Sprawdzamy jakich kolumn dotyczą braki i ile ich jest
columns=list(data.columns)
missing=[]
for col in columns: missing.append({col, data[col].isnull().sum()})
missing
```

```python
#Widzimy, że braki danych dotyczą jedynie zmiennych z 10 i 90 centylem zarobków
#Sprawdzamy, jakich rekordów dotyczą braki zmiennych
missing=data[data.isnull().any(axis=1)]
missing
```

```python
#Sprawdzamy ilość rekordów z brakami
len(missing.index)
```

```python
#Jak widzimy, jeśli w rekordzie występują braki danych, to są one obecne dla obu kolumn

#Najwięcej braków danych dotyczy kategorii Liberal Arts dla typu uniwersytetu
#Dla innych typów są to braki jednostkowe
#Sprawdzamy więc, jak wiele jest brakujących danych w stosunku do danych pełnych dla tego typu
len(data.loc[data['School Type'] == 'Liberal Arts'].index)
```

```python
len(missing.loc[missing['School Type'] == 'Liberal Arts'].index)
```

Najrozsądniejszym wyjściem wydaje się wykluczenie z analiz z udziałem zmiennych o 10 i 90 centylu zarobków wszystkich uniwersytetów z kategorii Liberal Arts lub przeprowadzanie ich jedynie na 13 pełnych rekordach z wyraźnym zaznaczeniem braków. Pozostałe braki dotyczą kategorii Engineering oraz Party, są na tyle nieliczne, że ich uzupełnienie nie powinno zaburzyć znacząco analiz.

## Uzupełnienie braków danych (z wyłączeniem Liberal Arts)

```python
'''Iterujemy kolejno po kolumnach w których występują braki danych
Używamy funkcji .loc, by wybrać jedynie rekordy, których typem nie jest "Liberal Arts"
Używamy funkcji .groupby, by wypełnić brakujące dane średnimi dla danego typu, używamy iteratora jako
indeksu wskazującego na odpowiednią kolumnę
W danej kolumnie uzupełniamy każdy brak danych średnią używając funkcji .mean
'''

for _ in ['Mid-Career 10th Percentile Salary', 'Mid-Career 90th Percentile Salary']:
    data.loc[data['School Type'] != 'Liberal Arts',_] = \
    data.groupby('School Type')[_].apply(lambda x:x.fillna(x.mean()))
```
