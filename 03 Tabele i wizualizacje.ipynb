```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#rozpakowywanie danych
data = pd.read_pickle("Data.pickle")
```

```python
#Dodanie kolumny z rÃ³Å¼nicÄ… zarobkÃ³w z poczÄ…tku i Å›rodka kariery
data["Difference Between Mid-Career and Starting Salary"] = data["Mid-Career Median Salary"] - data["Starting Median Salary"]
```

```python
data
```

# Tabele



## Czy poczÄ…tkowe Å›rednie zarobki absolwentÃ³w uczelni poszczegÃ³lnych rodzajÃ³w rÃ³Å¼niÄ… siÄ™ w zaleÅ¼noÅ›ci od regionu?

```python
region_start = data.filter(["Region", "School Type", "Starting Median Salary"], axis=1)
```

```python
region_start_sort = region_start.groupby(["School Type", "Region"]).median()
```

```python
region_start_sort
```

## KtÃ³ry region ma najwyÅ¼szÄ… Å›redniÄ… rÃ³Å¼nicÄ™ miÄ™dzy poczÄ…tkowymi a pÃ³Åºniejszymi zarÃ³bkami?

```python
zarobki = data.filter(["Region", "Difference Between Mid-Career and Starting Salary"], axis = 1)
```

```python
zarobki_sort = zarobki.groupby(["Region"]).median()
```

```python
zarobki_sort
```

## Jakie sÄ… najwyÅ¼sze i najniÅ¼sze mediany zarobkÃ³w na uczelniach poszczegÃ³lnych typÃ³w?

```python
imtired = data.groupby("School Type")["Mid-Career Median Salary"].agg(["max", "min"])
```

```python
imtired["Difference"] = imtired["max"] - imtired["min"]
```

```python
imtired_s = imtired.sort_values(by = "Difference", axis = 0, ascending = True)
```

```python
imtired_s
```

## Jak duÅ¼a jest rÃ³Å¼nica miÄ™dzy najwiÄ™cej i najmniej zarabiajÄ…cymi absolwentami szkÃ³Å‚ z poszczegÃ³lnych regionÃ³w?

```python
perce = data.groupby("Region")["Mid-Career 10th Percentile Salary", "Mid-Career 90th Percentile Salary"].median()
```

```python
perce["Difference"] = perce["Mid-Career 90th Percentile Salary"] - perce["Mid-Career 10th Percentile Salary"]
```

```python
perce_s = perce.sort_values(by = "Difference", axis = 0, ascending = True)
```

```python
perce_s
```

## Jak duÅ¼a jest rÃ³Å¼nica miÄ™dzy najwiÄ™cej i najmniej zarabiajÄ…cymi absolwentami szkÃ³Å‚ poszczegÃ³lnych typÃ³w?

```python
perce2 = data.groupby("School Type")["Mid-Career 10th Percentile Salary", "Mid-Career 90th Percentile Salary"].median()
```

```python
perce2["Difference"] = perce2["Mid-Career 90th Percentile Salary"] - perce2["Mid-Career 10th Percentile Salary"]
```

```python
perce2_s = perce2.sort_values(by = "Difference", axis = 0, ascending = True)
```

```python
perce2_s
```
