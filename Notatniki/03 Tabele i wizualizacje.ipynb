## Jakie sÄ… najwyÅ¼sze i najniÅ¼sze mediany zarobkÃ³w na uczelniach poszczegÃ³lnych typÃ³w?

```python
imtired = data.groupby("School Type")["Mid-Career Median Salary"].agg(["max", "min"])
```

### 02 PÅ‚aca w poÅ‚owie kariery wedÅ‚ug uczelni wyÅ¼szej

```python
def generuj_wykres(dataset, filename, opis):
    y = dataset["School Name"]
    labels = dataset["School Name"]

    fig = plt.figure(figsize=(15,6))

    x3 = dataset['Mid-Career 10th Percentile Salary']
    plt.scatter(x3, y, color='#bff0f8', label = '10 percentyl zarobkÃ³w ze Å›rodka kariery')
    plt.yticks(y, labels)

    x4 = dataset['Mid-Career 25th Percentile Salary']
    plt.scatter(x4, y, color='#6cb2ef', label = '25 percentyl (Q1) zarobkÃ³w ze Å›rodka kariery')

    x1 = dataset["Mid-Career Median Salary"]
    plt.scatter(x1, y, color='#1986e6', label='Mediana zarobkÃ³w ze Å›rodka kariery')

    x5 = dataset['Mid-Career 75th Percentile Salary']
    plt.scatter(x5, y, color='#8d6cef', label = '75 percentyl (Q3) zarobkÃ³w ze Å›rodka kariery')

    x6 = dataset['Mid-Career 90th Percentile Salary']
    plt.scatter(x6, y, color='#cdbff8', label = '90 percentyl zarobkÃ³w ze Å›rodka kariery')

    #x2 = dataset['Starting Median Salary']
    #plt.scatter(x2, y, color='#ef6c6c', marker="*", label='Mediana zarobkÃ³w z poczÄ…tku kariery')

    plt.xlabel('Roczne zarobki abolwentÃ³w [USD]')
    plt.title(opis)
    plt.legend(loc='lower right')
    plt.show()

    fig.savefig(filename, dpi=300)
```

```python
def generuj_wszystko(data, nazwa_wykresu, typ = None):
    mydata = data.sort_values(by = 'Mid-Career Median Salary', ascending=False)
    if typ != None: mydata = mydata[mydata['School Type'] == typ]
    mydata = mydata.head(20)
    mydata = mydata.sort_values(by = 'Mid-Career Median Salary', ascending=True)
    if typ != None: opis_wyk = 'Roczne zarobki absolwentÃ³w ' + str(len(mydata.index)) + ' uczelni typu '+ typ +'\n z najwyÅ¼szÄ… medianÄ… zarobkÃ³w z kariery'
    else: opis_wyk = 'Roczne zarobki absolwentÃ³w ' + str(len(mydata.index)) + ' uczelni \n z najwyÅ¼szÄ… medianÄ… zarobkÃ³w ze Å›rodka kariery'
    generuj_wykres(mydata, nazwa_wykresu, opis_wyk)
```

```python
generuj_wszystko(data,'wykres_2.png')
```
