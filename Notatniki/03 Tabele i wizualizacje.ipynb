# Tabele



## Czy początkowe średnie zarobki absolwentów uczelni poszczególnych rodzajów różnią się w zależności od regionu?

```python
region_start = data.filter(["Region", "School Type", "Starting Median Salary"], axis=1)
```

```python
region_start_sort = region_start.groupby(["School Type", "Region"]).median()
```

```python
region_start_sort
```

## Który region ma najwyższą średnią różnicę między początkowymi a późniejszymi zaróbkami?

```python
zarobki = data.filter(["Region", "Difference Between Mid-Career and Starting Salary"], axis = 1)
```

```python
zarobki_sort = zarobki.groupby(["Region"]).median()
```

```python
zarobki_sort
```

## Jakie są najwyższe i najniższe mediany zarobków na uczelniach poszczególnych typów?

```python
imtired = data.groupby("School Type")["Mid-Career Median Salary"].agg(["max", "min"])
```

```python
imtired["Difference"] = imtired["max"] - imtired["min"]
```

```python
imtired_s = imtired.sort_values(by = "Difference", axis = 0, ascending = True)
```

```python
imtired_s
```

## Jak duża jest różnica między najwięcej i najmniej zarabiającymi absolwentami szkół z poszczególnych regionów?

```python
perce = data.groupby("Region")["Mid-Career 10th Percentile Salary", "Mid-Career 90th Percentile Salary"].median()
```

```python
perce["Difference"] = perce["Mid-Career 90th Percentile Salary"] - perce["Mid-Career 10th Percentile Salary"]
```

```python
perce_s = perce.sort_values(by = "Difference", axis = 0, ascending = True)
```

```python
perce_s
```

## Jak duża jest różnica między najwięcej i najmniej zarabiającymi absolwentami szkół poszczególnych typów?

```python
perce2 = data.groupby("School Type")["Mid-Career 10th Percentile Salary", "Mid-Career 90th Percentile Salary"].median()
```

```python
perce2["Difference"] = perce2["Mid-Career 90th Percentile Salary"] - perce2["Mid-Career 10th Percentile Salary"]
```

```python
perce2_s = perce2.sort_values(by = "Difference", axis = 0, ascending = True)
```

### 02 Płaca w połowie kariery według uczelni wyższej

```python
def generuj_wykres(dataset, filename, opis):
    y = dataset["School Name"]
    labels = dataset["School Name"]

    fig = plt.figure(figsize=(15,6))

    x3 = dataset['Mid-Career 10th Percentile Salary']
    plt.scatter(x3, y, color='#bff0f8', label = '10 percentyl zarobków po 10 latach od rozpoczęcia kariery')
    plt.yticks(y, labels)

    x4 = dataset['Mid-Career 25th Percentile Salary']
    plt.scatter(x4, y, color='#6cb2ef', label = '25 percentyl (Q1) zarobków po 10 latach od rozpoczęcia kariery')

    x1 = dataset["Mid-Career Median Salary"]
    plt.scatter(x1, y, color='#1986e6', label='Mediana zarobków po 10 latach od rozpoczęcia kariery')

    x5 = dataset['Mid-Career 75th Percentile Salary']
    plt.scatter(x5, y, color='#8d6cef', label = '75 percentyl (Q3) zarobków po 10 latach od rozpoczęcia kariery')

    x6 = dataset['Mid-Career 90th Percentile Salary']
    plt.scatter(x6, y, color='#cdbff8', label = '90 percentyl zarobków po 10 latach od rozpoczęcia kariery')

    #x2 = dataset['Starting Median Salary']
    #plt.scatter(x2, y, color='#ef6c6c', marker="*", label='Mediana zarobków z początku kariery')

    plt.xlabel('Roczne zarobki abolwentów [USD]')
    plt.title(opis)
    plt.legend(loc='lower right')
    plt.show()

    fig.savefig(filename, dpi=300)
```

```python
def generuj_wszystko(data, nazwa_wykresu, typ = None):
    mydata = data.sort_values(by = 'Mid-Career Median Salary', ascending=False)
    if typ != None: mydata = mydata[mydata['School Type'] == typ]
    mydata = mydata.head(20)
    mydata = mydata.sort_values(by = 'Mid-Career Median Salary', ascending=True)
    if typ != None: opis_wyk = 'Roczne zarobki absolwentów ' + str(len(mydata.index)) + ' uczelni typu '+ typ +'\n z najwyższą medianą zarobków po 10 latach od rozpoczęcia kariery'
    else: opis_wyk = 'Roczne zarobki absolwentów ' + str(len(mydata.index)) + ' uczelni \n z najwyższą medianą zarobków po 10 latach od rozpoczęcia kariery'
    generuj_wykres(mydata, nazwa_wykresu, opis_wyk)
```

```python
generuj_wszystko(data,'wykres_2.png')
```

### 04 Wykresy kołowe

```python
regions = [len(data.loc[data['Region']=='California'].index), len(data.loc[data['Region']=='Midwestern'].index), len(data.loc[data['Region']=='Northeastern'].index), len(data.loc[data['Region']=='Southern'].index), len(data.loc[data['Region']=='Western'].index)]
labels = ['California', 'Midwestern', 'Northeastern', 'Southern', 'Western']
```

```python
fig1, ax1 = plt.subplots(figsize=(8,8))
ax1.pie(regions, labels=labels, autopct='%1.1f%%', startangle=90)
plt.rcParams.update({'font.size': 16})
ax1.axis('equal')
plt.show()

fig1.savefig('wykres_6.png', dpi=300)
```
